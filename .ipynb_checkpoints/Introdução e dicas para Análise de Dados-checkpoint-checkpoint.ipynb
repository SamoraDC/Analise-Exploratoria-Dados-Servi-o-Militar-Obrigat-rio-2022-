<h2> Uso de Gráficos </h2>

```python
# altera o tamanho do gráfico a ser gerado
pl.figure(figsize=(20,15))
# nomeia o dados do eixo x e y com os as colunas selecionadas do dataframe
pl.scatter(data=sm22, x="ESTADO_CIVIL", y="ZONA_RESIDENCIAL");
```

```python
# gerar um gráfico de histograma
# aqui podemos ver a realidade Brasileira desde o exôdo rural com 85% da população vivendo em zona urbana
pl.hist(data=sm22, x="ZONA_RESIDENCIAL");
```

```python
# gráfico histograma com a coluna de dados "ESTADO_CIVIL"
# onde se vê a magnitude de convocados que são solteiros
# algo perfeitamente normal devido a baixa idade dos convocados
pl.figure(figsize=(16,10))
pl.hist(data=sm22, x="ESTADO_CIVIL");
```

```python
# cria um gráfico seaborn de contagem da coluna "DISPENSA" 
sb.countplot(data=sm22, x="DISPENSA");
```

```python
# aqui nós alteramos um pouco a o nosso gráfico, e seguimos as boas práticas de análise exploratória,
# no sentido de remover o que não for estritamente necessário, inclusive as linhas

# criar contagem
dispensa_contagem = sm22["DISPENSA"].value_counts()

# tamanho do gráfico
pl.figure(figsize=(15,10))

# gráfico de barras com índice e contagem 
pl.bar(
    dispensa_contagem.index, # valor eixo x
    dispensa_contagem.values, # valor eixo y
    color=["lightblue", "darkred"]
)
# informações e tamanho das fontes
pl.xlabel("SITUAÇÃO", fontsize=14) # nome do eixo e tamanho da fonte eixo x
pl.ylabel("CONTAGEM", fontsize=14) # nome do eixo e tamanho da fonte eixo y   
# título e tamanho da fonte
pl.title("DISTRIBUIÇÃO DE DISPENSA", fontsize=16)
# retirar as linhas
pl.gca().spines["top"].set_visible(False) # retira as linhas de cima 
pl.gca().spines["right"].set_visible(False) # retira as linhas da direita
# apresenta o gráfico
pl.show;
```

```python
# Criação de gráfico de dispersão
# com uma correlação muito fraca entre o "PESO" e "CINTURA"
fig,ax=pl.subplots(figsize=(25,20))
ax.scatter(sm22["PESO"], sm22["CINTURA"])
# setar informações do gráfico e tamanho das fontes
ax.set_title("PESO vs CINTURA", fontsize=24),# nome do título e tamanho da fonte
ax.set_xlabel("PESO", fontsize=20),# nome do eixo x e tamanho da fonte
ax.set_ylabel("CINTURA", fontsize=20),# nome do eixo y e tamanho da fonte
# remoção das linhas 
ax.spines["top"].set_visible(False),# retira a coluna de cima
ax.spines["right"].set_visible(False),# retira a coluna da direita
# apresenta o gráfico
pl.show()
```

```python
# ciriação de mais um gráfico de dispersão 
pl.scatter(data=sm22, x="CABECA", y="ALTURA")
```

```python
# criação de 2 gráficos de dispersão
# criar o local e configuração do plot
fig, axes = pl.subplots(
    1, # número de linhas
    2, # número de colunas
    figsize=(12,8) # tamanho do plt
)
# gráfico de dispersão na posição 0 (linha 0, coluna 0)
axes[0].scatter(sm22["PESO"], sm22["CINTURA"])
axes[0].set_xlabel("PESO")
axes[0].set_ylabel("CINTURA")
axes[0].set_title("PESO vs CINTURA")
# gráfico de dispersão na posição 1 (linha 0, coluna 1)
axes[1].scatter(sm22["CALCADO"], sm22["CINTURA"])
axes[1].set_ylabel("CALCADO")
axes[1].set_xlabel("CINTURA")
axes[1].set_title("CALCADO vs CINTURA")
# ajusta espaço entre plots
pl.tight_layout()
# apresenta o gráfico
pl.show()
```
